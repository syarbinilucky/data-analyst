```python
# Mengimpor library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("netflix_titles.csv")
# print(type(df))
df.head(5)
```

```python
df = df.drop_duplicates()
df.head(3)
```

```python
df = df.dropna()
df.count()
```

```python
df = df.drop(['show_id', 'description'], axis=1)

df.head(3)
```

```python
# 1. Distribusi Data Tipe Konten
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='type', palette='viridis')
plt.title('Distribusi Tipe Konten di Netflix')
plt.xlabel('Tipe Konten')
plt.ylabel('Jumlah')
plt.show()
```

```python
# Mengonversi kolom 'date_added' ke tipe datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Memastikan tidak ada nilai kosong setelah konversi
print(df['date_added'].isnull().sum(), "nilai null ditemukan di 'date_added'")

# Mengambil tahun dari kolom 'date_added'
df['release_year'] = df['date_added'].dt.year

# Memvisualisasikan tren rilis konten
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='release_year', palette='viridis')
plt.title('Tren Rilis Konten di Netflix')
plt.xlabel('Tahun Rilis')
plt.ylabel('Jumlah Konten')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Analisis Genre dan Negara Asal
# Menghitung genre terpopuler
genre_counts = df['listed_in'].str.split(', ').explode().value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='viridis')
plt.title('10 Genre Terpopuler di Netflix')
plt.xlabel('Jumlah Konten')
plt.ylabel('Genre')
plt.show()
```

```python
# Menghitung negara terpopuler
country_counts = df['country'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.values, y=country_counts.index, palette='viridis')
plt.title('10 Negara Terpopuler di Netflix')
plt.xlabel('Jumlah Konten')
plt.ylabel('Negara')
plt.show()
```

```python
# 4. Durasi atau Rating Konten
# Menghitung rating terpopuler
rating_counts = df['rating'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='viridis')
plt.title('Distribusi Rating Konten di Netflix')
plt.xlabel('Rating')
plt.ylabel('Jumlah Konten')
plt.xticks(rotation=45)
plt.show()
```

```python
# Menghitung durasi rata-rata film
df['duration'] = df['duration'].str.replace(' min', '', regex=False)  # Menghapus ' min'
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')  # Mengubah ke float, mengganti kesalahan dengan NaN

# Menghitung durasi rata-rata hanya untuk film
average_duration = df[df['type'] == 'Movie']['duration'].mean()
print(f'Rata-rata durasi film di Netflix: {average_duration} menit')
```

```python
# 5. Data Missing Values
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
plt.figure(figsize=(12, 6))
sns.barplot(x=missing_values.index, y=missing_values.values, palette='viridis')
plt.title('Jumlah Data Missing di Setiap Kolom')
plt.xlabel('Kolom')
plt.ylabel('Jumlah Missing Values')
plt.xticks(rotation=45)
plt.show()
```
